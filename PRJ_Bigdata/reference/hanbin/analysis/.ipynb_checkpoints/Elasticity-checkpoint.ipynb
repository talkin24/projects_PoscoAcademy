```python
df_pr_mean_year = pd.DataFrame([df_pr['subscribe_price'].groupby(df_pr['price_year']).mean()])
df_pr_mean_year
```

```python
df_dl_year = pd.DataFrame([df_dl['down_year'].value_counts()]).transpose()
df_dl_year
```

```python
num_register = pd.DataFrame([df_cs['register_year'].value_counts()])
num_register_cum = num_register.cumsum(axis = 1)
num_register_cum
```

```python
num_register_cum.transpose()
```

```python
num_drop = pd.DataFrame([df_cs['drop_year'].value_counts()[1:]])
num_drop
```

```python
num_custom_year = []
num_custom_year.append(int(num_register_cum[2014]))


num_custom_year.append(int(num_register_cum[2015]) - int(num_drop['2015']))
num_custom_year.append(int(num_register_cum[2016]) - int(num_drop['2016']))
num_custom_year.append(int(num_register_cum[2017]) - int(num_drop['2017']))
num_custom_year.append(int(num_register_cum[2018]) - int(num_drop['2018']))

num_custom_year
```

```python
df_dl_year = df_dl_year.sort_index()
df_dl_year
```

```python
df_dl_year['num_custom'] = num_custom_year
df_dl_year
```

```python
df_dl_year['per_capita_down'] = df_dl_year['down_year']/df_dl_year['num_custom']
df_dl_year['price_mean'] = df_pr_mean_year.transpose()
df_dl_year.drop(['price_year'],axis=1)
df_dl_year
```

```python
d_price_mean = [0]
for i in range(2015,2019):
    d_price_mean.append(df_dl_year['price_mean'][i] - df_dl_year['price_mean'][i-1])
    
d_per_capita_down = [0]
for i in range(2015,2019):
    d_per_capita_down.append(df_dl_year['per_capita_down'][i] - df_dl_year['per_capita_down'][i-1])

df_dl_year['d_price_mean'] = d_price_mean
df_dl_year['d_per_capita_down'] = d_per_capita_down
df_dl_year
```

```python
elstc = [0]
for i in range(2015,2019):
    elstc.append((df_dl_year['d_per_capita_down'][i] / df_dl_year['per_capita_down'][i])/(df_dl_year['d_price_mean'][i] / df_dl_year['price_mean'][i]))
df_dl_year['elstc'] = elstc
df_dl_year
```
