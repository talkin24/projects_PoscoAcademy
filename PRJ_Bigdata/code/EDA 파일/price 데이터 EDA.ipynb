```python
df_download=pd.read_csv('movie_download_cleaned.csv',engine='python')
df_download.head()
```

```python
df_customer2=pd.read_csv('movie_customer_cleaned_ver4.csv',engine='python')
df_customer2.head()
```

```python
sum(df_download[df_download.customer_id=='C5001']['down_price'])
```

```python
for i in range(len(df_customer2)):
    df_customer2.loc[i,'down_price']=sum(df_download[df_download.customer_id==df_customer2.customer_id[i]]['down_price'])
```

```python
df_customer2.head()
```

```python
df_customer2['down_price']=df_customer2['down_price'].round(0).astype('int')
```

```python
import matplotlib.pylab as plt

plt.rcParams["figure.figsize"] = (14,10)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 

sns.barplot(x='down_price',y='state',data=df_customer2,estimator=np.sum)
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 

sns.barplot(x='age_group',y='down_price',data=df_customer2,estimator=np.sum)
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 

sns.barplot(x='gender',y='down_price',data=df_customer2,estimator=np.sum)
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 

sns.barplot(x='married',y='down_price',data=df_customer2,estimator=np.sum)
```

```python
sns.pairplot(x_vars='down_price',y_vars='activation_day',data=df_customer2)
```

```python
sns.pairplot(x_vars='down_price',y_vars='download_cnt',data=df_customer2)
```

```python
for i in range(len(df_customer2)):
    df_customer2.loc[i,'down_price/cnt']=df_customer2.down_price[i]/df_customer2.download_cnt[i]
```
