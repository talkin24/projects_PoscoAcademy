# release_year 이상치 확인 및 처리

```python
import re
p = re.compile(r"^\d{4}$")
for i in range(len(df_final_meta["release_year"])):
    if len(re.findall(p, df_final_meta["release_year"][i]))==0 : 
        df_final_meta.loc[i,"release_year"] = df_final_meta.loc[i,"release_year"][:4]
```

# release_date 이상치/결측치

```python
df_final_meta[df_final_meta["release_date"].isna()]
```

```python
mth = {"Jan":"01","Feb":"02", "Mar":"03", "Apr":"04", "May":"05","Jun":"06","Jul":"07","Aug":"08",
       "Sep":"09", "Oct":"10", "Nov":"11","Dec":"12", 
       "January":"01","February":"02", "March":"03", "April":"04", "May":"05","June":"06","July":"07",
       "August":"08", "September":"09", "October":"10", "November":"11","December":"12"}


p = re.compile(r"^\d{2} (Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{4}$")
count = 0
for i in range(len(df_final_meta["release_date"])):
        if len(re.findall(p,df_final_meta["release_date"][i]))==0 : 
            
            if df_final_meta["release_date"][i][:6]=="(Wide)":
                df_final_meta.loc[i, "release_date"] = df_final_meta.loc[i, "release_date"][6:]
                
            if df_final_meta["release_date"][i]!="." :
                L = df_final_meta["release_date"][i].split()
                month = mth[L[0]]; year = L[2]
                if L[1][:2].isnumeric() : 
                    day = L[1][:2]
                else : day = "0"+L[1][:1]
                df_final_meta.loc[i, "release_date"] = month+"-"+day+"-"+year
                      
        else : 
            L = df_final_meta["release_date"][i].split() # 잘드간 애들
            df_final_meta.loc[i, "release_date"] = mth[L[1]]+"-"+L[0]+"-"+L[2]
        print(df_final_meta.loc[i,"release_date"])
```

# awards 파생변수들


```python
df_final_meta[df_final_meta["awards"].isna()]
```

```python
import re
import math

# print(math.isnan(df_meta["awards"][12]))
count =0
big_awards = []
small_win = []
small_nom = []

for x in df_final_meta["awards"]:
    
    if x ==".":
        big_awards.append(0)
        small_win.append(0); small_nom.append(0)
        count+=1

    elif re.findall(re.compile(r"^\d+ win[s]?.$"),x):
        big_awards.append(0); small_nom.append(0)
        small_win.append(re.split(re.compile(r"[\s]"),x)[0])
        count+=1
    elif re.findall(re.compile(r"^\d+ nomination[s]?.$"),x):
        big_awards.append(0); small_win.append(0)
        small_nom.append(int(re.split(re.compile(r"[\s]"),x)[0]))
        count+=1
    elif re.findall(re.compile(r"^\d+ win[s]? & \d+ nomination[s]?.$"),x):
        L = re.split(re.compile(r"[\s]"),x)
        big_awards.append(0)
        small_win.append(int(L[0])); small_nom.append(int(L[3])+int(L[0]))
        count+=1

    elif re.findall(re.compile(r"^(Won|Nominated for) \d+ [a-zA-Z0-9\s]+. Another \d+ win[s]? & \d+ nomination[s]?.$"), x):
        L = re.split(re.compile(r"[\s]"),x)
        big_awards.append(1)
        small_win.append(int(L[-5])); small_nom.append(int(L[-2])+int(L[-5]))
        count+=1
    elif re.findall(re.compile(r"^(Won|Nominated for) \d+ [a-zA-Z0-9\s]+. Another \d+ win[s]?.$"),x):
        L = re.split(re.compile(r"[\s]"),x)
        big_awards.append(1)
        small_win.append(int(L[-2])); small_nom.append(int(L[-2]))
        count+=1
    elif re.findall(re.compile(r"^(Won|Nominated for) \d+ [a-zA-Z0-9\s]+. Another \d+ nomination[s]?.$"),x):
        L = re.split(re.compile(r"[\s]"),x)
        big_awards.append(1)
        small_win.append(0); small_nom.append(int(L[-2]))
        count+=1

    elif re.findall(re.compile(r"^(Won|Nominated for) \d+ [a-zA-Z0-9\s]+.$"),x):
        L = re.split(re.compile(r"[\s]"),x)
        big_awards.append(1)
        small_win.append(0); small_nom.append(0)
        count+=1
print(count)
```

```python
df_final_meta["big_awards"] = big_awards
df_final_meta["awards_win"] = small_win
df_final_meta["awards_nomin"] = small_nom
```
